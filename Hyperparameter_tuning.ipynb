```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor()
```

```python
kfold = KFold(n_splits=5, shuffle=True,random_state=1)
scores=cross_val_score(knn, X,y, cv=kfold, scoring = 'r2')
```

```python
scores.mean()
```

GridSearch CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn = KNeighborsRegressor()
```

```python
param_grid={
    'n_neighbors' :[1,3,5,7,10,12,15,17,20], # 1 to 20 added alternate number
    'weights':['uniform', 'distance'],
    'algorithm':['ball_tree','kd_tree','brute'],
    'p':[1,2] # 1 for "Manhattan", 2 for "Euclidean distance"
}

# This is 4d Grid and here we have "108" Unique combination (9 neigbour , 3 algorithm, 2 distance, 2 weights 9*3*2*2)
# SO GrisSearchCV WILL RUN 108 INDEPENDENT KNN
```

```python
gcv = GridSearchCV(knn, param_grid,scoring='r2',refit=True,cv=kfold, verbose=2)
# Now when apply CV(cross validation)kfold=5 upon 108 unique knn it will trained 540 different models.
# refit = True (it means the best parameter which find will use to train model again so that we can do prediction)
# cv= will provide how much kfold we want
# verbose = will show the result
```

```python
gcv.fit(X,y) # now fitting gridcv on data
```

```python
gcv.best_params_. # it give best parameter value
```

```python
gcv.best_score_ # this is the best r2 score we get ".61" which is significant improvement as before grid cv we get r2score ".47"
```

```python
gcv.cv_results_ # this will show overall result in dictionary
```
